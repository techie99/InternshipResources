```python
import pandas as pd
data = pd.read_csv("Weather Data (1).csv")
print(data)
```

```python
print(data.head())
```

```python
#Find all unqiue "wind speed" values in the data
unique_wind_speed = data["Wind Speed_km/h"].unique()
b=unique_wind_speed.tolist()
print(len(b))
```

```python
#Find all the number of times when the "weather is exactly clear"
clear_weather = data[data["Weather"] == "Clear"]
#clear_weather = data[data.Weather == "clear"]
print(clear_weather.value_counts())
#print(data[data.Weather=="Clear"])
#print(data["Weather"].value_counts())
```

```python
#Find the number of times when the "wind speed was exactly 4 km/hr"
wind_speed_4 = data[data["Wind Speed_km/h"] == 4]
print(len(wind_speed_4))
#or
#data[data["Wind Speed_km/h"] == 4].shape[0]
```
