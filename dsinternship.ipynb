# Pandas

```python
import pandas as pd
data = pd.read_csv("marriage_divorce_india_with_id.csv")
print(data)
```

```python
#1. Compare the average income levels between love and arranged marriages.
average_income = data.groupby('Marriage Type')['Income Level (INR per month)'].mean()

print(average_income)
```

```python
#2. Identify the top 3 education levels with the highest divorce rates.
top_education_levels = data[data['Divorce Status'] == 'Yes']['Education Level'].value_counts().head(3)
print(top_education_levels)
```

```python
#3. Determine the average marriage duration based on the number of children.
average_duration_children = data.groupby('Children')['Marriage Duration (Years)'].mean()
print(average_duration_children)
```

```python
#4. Analyze the influence of family involvement on divorce rates.

data['Family Involvement Numeric'] = pd.factorize(data['Family Involvement'])[0]
data['Divorce Numeric'] = (data['Divorce Status'] == 'Yes').astype(int)

correlation = data['Family Involvement Numeric'].corr(data['Divorce Numeric'])

print(f"Correlation between Family Involvement and Divorce Rates: {correlation:.2f}")
```

```python
#5. Find the most common age range for divorce in urban and rural areas.
bins = [18, 25, 35, 45, 60]  # Define age bins
labels = ['18-24', '25-34', '35-44', '45-60']  # Labels for age ranges
data['Age Range'] = pd.cut(data['Age at Marriage'], bins=bins, labels=labels)
age_divorce_urban_rural = data[data['Divorce Status'] == 'Yes'].groupby(['Urban/Rural', 'Age Range']).size().unstack()
print(age_divorce_urban_rural)
```

```python
#6. Calculate the proportion of love vs. arranged marriages by caste/religion.
marriage_proportions = data.groupby('Caste/Religion')['Marriage Type'].value_counts(normalize=True).unstack() * 100
print(marriage_proportions)
```

```python
#7. Identify the top 3 income brackets with the longest marriage durations.
data['Income Bracket'] = pd.qcut(data['Income Level (INR per month)'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])
top_income_brackets = data.groupby('Income Bracket')['Marriage Duration (Years)'].mean().sort_values(ascending=False).head(3)
print(top_income_brackets)
```

```python
#8. Examine how urban/rural backgrounds correlate with education levels.
urban_rural_education = data.groupby(['Urban/Rural', 'Education Level']).size().unstack()
print(urban_rural_education)
```

```python
#9. Explore the relationship between age at marriage and the likelihood of divorce.
age_divorce_likelihood = data.groupby('Divorce Status')['Age at Marriage'].mean()
print(age_divorce_likelihood)
```

```python
#10. Highlight the most common family involvement levels in divorced marriages.
family_involvement_divorced = data[data['Divorce Status'] == 'Yes']['Family Involvement'].value_counts()
print(family_involvement_divorced)
```

# Seaborn


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.swarmplot(x='Marriage Type', y='Age at Marriage', data=data, palette='Set1')
plt.title('Age at Marriage vs. Marriage Type')
plt.show()
```

```python
#2. Create a pairplot for numerical variables grouped by divorce status.
numerical_columns = ['Age at Marriage', 'Income Level (INR per month)', 'Marriage Duration (Years)']
sns.pairplot(data, vars=numerical_columns, hue='Divorce Status', palette='coolwarm', diag_kind='kde')
plt.show()
```

```python
#3. Visualize the distribution of family involvement levels with a violin plot.
sns.violinplot(x='Family Involvement', y='Marriage Duration (Years)', data=data, palette='muted')
plt.title('Distribution of Family Involvement Levels')
plt.show()

```

```python
#4. Use a heatmap to highlight correlations between age, income, and marriage duration.
correlation_data = data[['Age at Marriage', 'Income Level (INR per month)', 'Marriage Duration (Years)']].corr()
sns.heatmap(correlation_data, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#5Show the density distribution of marriage durations using kdeplot.
sns.kdeplot(data['Marriage Duration (Years)'], shade=True, color='purple')
plt.title('Density Distribution of Marriage Durations')
plt.xlabel('Marriage Duration (Years)')
plt.show()
```
